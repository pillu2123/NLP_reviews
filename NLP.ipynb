```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\user\Desktop\machine learning\Machine Learning A-Z (Codes and Datasets)\Part 7 - Natural Language Processing\Section 36 - Natural Language Processing\Python\Restaurant_Reviews.tsv",delimiter="\t",quoting=3)
dataset.head(3)
```

```python
import re
import nltk
nltk.download("stopwords") # to remove words like "is" "a" "the"
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer # to get the stem of the root word, like loved will become love
corpus=[]
for i in range(dataset.shape[0]):
    review=re.sub('[^a-zA-Z]',' ',dataset["Review"][i]) # removing commas dots anything xcept alphabets
    review=review.lower()
    review=review.split()
    ps=PorterStemmer()
    all_words=stopwords.words("english")
    all_words.remove('not')
    review=[ps.stem(word) for word in review if not word in set(all_words)] # stemming of the words 
    review=" ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer() # to tokenize the words in review
X=cv.fit_transform(corpus).toarray() 
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

# Svc

```python
from sklearn.svm import SVC
classifier=SVC(kernel="rbf",random_state=42)
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(confusion_matrix(y_test,y_pred))
accuracy_score(y_test,y_pred)
```

# Naive_bayes classifier 

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(confusion_matrix(y_test,y_pred))
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=100)
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```
